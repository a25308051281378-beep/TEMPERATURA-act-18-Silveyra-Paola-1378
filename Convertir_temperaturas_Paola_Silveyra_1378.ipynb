#### Fórmulas de Conversión
Para entender cómo pasar de una escala a otra, utilizamos las siguientes relaciones matemáticas:De Celsius a Fahrenheit:$$F = (C \times \frac{9}{5}) + 32$$(Multiplicamos por 1.8 y sumamos 32)De Fahrenheit a Celsius:$$C = (F - 32) \times \frac{5}{9}$$(Restamos 32 y multiplicamos por 0.555...)
#### Ejemplos Resueltos Paso a Paso
Ejemplo 1: Convertir 25 °C a °FIdentificar la fórmula:

$F = (C \times 1.8) + 32$

Sustituir el valor: $F = (25 \times 1.8) + 32$

Multiplicar: $25 \times 1.8 = 45$

Sumar: $45 + 32 = 77$

Resultado: 25 °C equivalen a 77 °F.

Ejemplo 2: Convertir 98.6 °F a °CIdentificar la fórmula:

$C = (F - 32) / 1.8$

Sustituir el valor:

$C = (98.6 - 32) / 1.8$

Restar: $98.6 - 32 = 66.6$

Dividir: $66.6 / 1.8 = 37$

Resultado: 98.6 °F equivalen a 37 °C (temperatura corporal promedio).

```python
# Funcion para convertir de celsius a fahrennheit
def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

# Funcion para convertir de fahrennheit a celsius
def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

# Funcion principal del menu
def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")

        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("Paola Silveyra NL = 50: ")
            break
        else:
            print("Opción no válida, intenta de nuevo.")

# Llamada principal para iniciar el programa
if __name__ == "__main__":
    menu()
```
